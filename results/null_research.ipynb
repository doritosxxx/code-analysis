```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
```

```python
file_name = "res.csv"
link = "https://raw.githubusercontent.com/doritosxxx/code-analysis/main/results/all.compressed.csv"
def read_csv_file():
    return pd.read_csv(link, sep=",", dtype='unicode')
```

```python
data = read_csv_file()
data = data.replace('-', np.NaN)
data = data[["acoc",'loc',"coc", "mxcoc", "mncoc" , "cc", "nullReferences"]].astype(float)
```

```python
print(data.head())
print(data.info())
data.corr(numeric_only = False)
```

```python
data_norm = data
data_norm = data_norm[~data_norm['loc'].isnull()]
data_norm = data_norm[data_norm['loc']>0]
data_norm["nullReferences"] = data_norm["nullReferences"]
print(data_norm.head())
print(data_norm.describe())
print(data_norm.info())

data_norm[["nullReferences",]] = data_norm[['nullReferences']].div(data_norm["loc"], axis=0)
print(data_norm.head())
print(data_norm.describe())
print(data_norm.info())
data_norm.corr(numeric_only = False)
```
